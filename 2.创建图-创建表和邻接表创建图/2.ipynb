```python
import matplotlib.pyplot as plt
import pandas as pd
import networkx as nx
plt.rcParams['axes.unicode_minus'] = False  # 正常显示负号
```

### 导入三元组连接表
##### 数据来源：OpenKG-四大名著人物关系知识图谱和OWL本体：http://openkg.cn/dataset/ch4masterpieces

```python
df = pd.read_csv('/kaggle/input/test-data/triples.csv')
df
```

### 通过连接表Edge List创建图

```python
G = nx.DiGraph()
```

```python
edges = [edge for edge in zip(df['head'],df['tail'])]
# print(edges)
G.add_edges_from(edges)
# nx.draw(G)
```

```python
G.edges('关羽')
```

### 可视化

```python
# 节点排版布局-默认弹簧布局
pos = nx.spring_layout(G,seed=123)

plt.figure(figsize=(15,15))
nx.draw(G,pos=pos) # 中文无法正常显示 with_labels=True
```

### 保存并载入邻接表Adjacency List

```python
for line in nx.generate_adjlist(G):
    print(line)
```

```python
# 将邻接表导出为本地文件 grid.edgelist
nx.write_edgelist(G,path="grid.edgelsit",delimiter=":")
```

```python
# 从本地文件grid.edgelist 读取邻接表
H = nx.read_edgelist(path="grid.edgelsit",delimiter=":")
```

```python
# 可视化
plt.figure(figsize=(15,14))
pos = nx.spring_layout(H,iterations=3,seed=5)
nx.draw(H,pos)
plt.show()
```
