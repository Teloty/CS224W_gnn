### 创建连接
在NetworkX中创建连接，设置连接的属性特征。

```python
import networkx as nx
```

### 创建图
NetworkX支持有向图 （directed graph）、无向图 （undirected graph）、带权重的图（weighte graph）、多路图 （multigraph）。\
文档: https://networkx.org/documentation/stable/reference/classes/index.html

```python
# 创建无向图
G = nx.Graph()
print(G.is_directed())
```

```python
# 给整张图添加属性特征
G.graph['Name'] = 'HelloWorld'
print(G.graph)
```

```python
# 创建有向图
H = nx.DiGraph()
print(H.is_directed())
```

### 创建单个节点

```python
G.add_node(0,feature=5,lable=0,zihao=2)
```

```python
G.nodes[0]
```

### 创建多个节点

```python
G.add_nodes_from([
    (1,{'feature':1,'label':2,'zihao':3}),
    (2,{'feature':2,'label':2,'zihao':4})
])
```

### 全图节点信息

```python
G.number_of_nodes()
```

```python
G.nodes()
```

```python
G.nodes(data=True)
```

```python
# 遍历所有节点，data=True 表示输出节点特征属性信息

for node in G.nodes(data=True):
    print(node)
```

### 创建单个连接，设置属性特征
特征属性的名字可以随便起

```python
G.add_edge(0,1,weight=0.5,like=3)
```

### 创建多个连接

```python
G.add_edges_from([
    (1,2,{'weight':0.3,'like':5}),
    (2,0,{'weight':0.1,'like':8})
])
```

```python
G.edges[(0,1)]
```

### 可视化

```python
nx.draw(G,with_labels = True)
```

### 全图连接信息

```python
G.number_of_edges()
```

```python
G.size()
```

```python
G.edges()
```

```python
G.edges(data=True)
```

```python
# 遍历所有节点，data=True 表示输出连接特征属性信息

for edge in G.edges(data=True):
    print(edge)
```

### 节点的连接数(Node Degree)

```python
# 指定节点
node_id = 1
```

```python
G.degree[node_id]
```

```python
# 指定节点的所有相邻节点
for neighbor in G.neighbors(node_id):
    print("Node {} has neighbor {}".format(node_id,neighbor))
```
