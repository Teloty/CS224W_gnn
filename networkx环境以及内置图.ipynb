## 社交网络

```python
# 空手道俱乐部数据集
G_club = nx.karate_club_graph()
nx.draw(G_club,with_labels=True)
```

```python
G_club.nodes[5]["club"]
```

```python
G_club.nodes[9]["club"]
```

```python
# 雨果《悲惨世界》人物关系
G_novel = nx.les_miserables_graph()
plt.figure(figsize=(12,10))
pos = nx.spring_layout(G_novel,seed=10)
nx.draw(G_novel,pos,with_labels=True)
```

```python
# Florentine families graph
G_family = nx.florentine_families_graph()
nx.draw(G_family,with_labels=True)
```
